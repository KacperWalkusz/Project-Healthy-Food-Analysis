```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

source : https://world.openfoodfacts.org/data

```python
df= pd.read_csv('en.openfoodfacts.org.products.csv', sep="\t", on_bad_lines='skip', low_memory=False)
```

```python
df
```

```python
df.columns.values
```

```python
df_clean = df.dropna(subset=['product_name'])
```

```python
df_clean =  df_clean.dropna(subset=['brands'])
```

```python
df_clean
```

```python
min_completed = len(df) * 0.1
df_clean = df_clean.dropna(thresh=min_completed, axis=1)
```

```python
df_clean
```

```python
cols_to_be_deleted = ['code', 'url', 'creator', 'created_t', 'created_datetime',
       'last_modified_t', 'last_modified_datetime', 'last_modified_by',
       'last_updated_t', 'last_updated_datetime','quantity', 'packaging', 'packaging_tags', 'packaging_en','ingredients_text',
       'ingredients_tags', 'serving_size','serving_quantity','pnns_groups_1', 'pnns_groups_2', 'food_groups',
       'food_groups_tags', 'food_groups_en', 'states', 'states_tags',
       'states_en', 'product_quantity', 'unique_scans_n',
       'popularity_tags', 'completeness', 'last_image_t',
       'last_image_datetime', 'main_category', 'main_category_en',
       'image_url', 'image_small_url', 'image_ingredients_url',
       'image_ingredients_small_url', 'image_nutrition_url',
       'image_nutrition_small_url']
```

```python
df_clean = df_clean.drop(columns=cols_to_be_deleted)
```

```python
df_clean
```

```python
df_clean.columns.values
```

```python
df_clean.to_csv('en.openfoodfacts.org.products_clean.csv', sep='\t', index=False)
```
